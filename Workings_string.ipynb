```python
%%sql

select concat(emp_code,'-',name)
from new_employees;
```

```python
%%sql
SELECT name, LENGTH(name) AS name_length
FROM new_employees;
```

```python
%%sql
SELECT
    name,
    UPPER(name) AS upper_name,
    LOWER(name) AS lower_name
FROM new_employees;
```

```python
%%sql
SELECT name, SUBSTRING(name, 1, 5) AS name_prefix
FROM new_employees;
```

```python
%%sql
SELECT CONCAT('  ', name, '  ') AS padded_name,
TRIM(CONCAT('  ', name, '  ')) AS trimmed_name
FROM new_employees;
```

```python
%%sql
SELECT name, REPLACE(name, 'Johnson', 'John') AS new_name
FROM new_employees;
```

```python
%%sql
SELECT email, LOCATE('@', email) AS at_position
FROM new_employees;
```

```python
%%sql
SELECT CONCAT_WS(' - ', name, email, department_id) AS contact_info
FROM new_employees;
```
